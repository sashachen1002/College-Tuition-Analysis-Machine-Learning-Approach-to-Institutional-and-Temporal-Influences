```python
# Read CollegeTuition.csv into a data frame called collegeTuition.
collegeTuition = pd.read_csv('CollegeTuition.csv')
```

```python
# Check the data type of the 'Year' column
print(collegeTuition['Year'].dtype)

# If 'Year' is not a string type, convert it to string first
collegeTuition['Year'] = collegeTuition['Year'].astype(str)

# Then split by hyphen and convert to integer
collegeTuition['Year'] = collegeTuition['Year'].str.split('-').str[0].astype(int)
# This code first converts the 'Year' column to a string type, then splits it at the hyphen, and takes the first part of the split
```

```python
#print the first 5 rows of the data frame
print(collegeTuition.head())
```

```python
# Filter out 'All Institutions'
filtered_collegeTuition = collegeTuition[collegeTuition['Institution Type'] != 'All Institutions']

# Display the first ten records of the collegeTuition data frame.
filtered_collegeTuition.head(10)
```

```python
# Convert 'Institution Type' into a binary variable
filtered_collegeTuition['Institution_Type_Binary'] = filtered_collegeTuition['Institution Type'].apply(lambda x: 1 if x == 'Private Institutions' else 0)
filtered_collegeTuition = filtered_collegeTuition.drop(['Institution Type'], axis=1)

# Convert 'School Type' into a binary variable
filtered_collegeTuition['School_Type_Binary'] = filtered_collegeTuition['School Type'].apply(lambda x: 1 if x == '4-year' else 0)
filtered_collegeTuition = filtered_collegeTuition.drop(['School Type'], axis=1)


# Confirm that 'Institution Type' is converted into a binary variable and display the first ten records of the collegeTuition data frame.
filtered_collegeTuition.head(10)
```

```python
# Display the structure of collegeTuition to the console.
filtered_collegeTuition.info() 
```

```python
collegeTuitionTraining, collegeTuitionTesting = train_test_split(filtered_collegeTuition, test_size=0.25, random_state=42)
```

```python
collegeTuitionTraining['Institution_Type_Binary'].value_counts()
```

```python
collegeTuitionModel = LogisticRegression().fit(filtered_collegeTuition.drop('Institution_Type_Binary', axis=1), filtered_collegeTuition['Institution_Type_Binary'])
```

```python
# Display the logistic regression model coefficients
collegeTuitionModel.coef_
```

```python
# 23. Calculate the odds ratios for each of the 7 independent variable coefficients.
np.exp(collegeTuitionModel.coef_)

#在邏輯回歸中，這個比率是指當特徵值增加一單位時，正類事件發生概率與負類事件發生概率的比值。
#勝算比大於1表示特徵與正類的關聯性增加。例如，勝算比為1.08表示特徵增加一單位時，正類的勝算比會增加8%。
#勝算比小於1但大於0表示特徵與負類的關聯性增加。例如，勝算比為0.99表示特徵增加一單位時，正類的勝算比會減少1%。
#勝算比等於1表示該特徵對於預測結果沒有影響。
#在您的例子中，所有特徵的勝算比都接近於1，這意味著這些特徵對於預測是否為私立學院的影響都相對較小。這可能表明模型中的特徵對於區分私立與非私立學院的能力有限，或者這些特徵和目標變數之間的關聯非常微弱。这也可能是因为这些特征已经在某种程度上被标准化，因此其系数表示的是标准化后单位变化的影响。
```

```python
collegeTuitionTesting['PredictedInstitution_Type_Binary'] = collegeTuitionModel.predict(collegeTuitionTesting.drop('Institution_Type_Binary', axis=1))
collegeTuitionTesting['PredictedInstitution_Type_Binary'] = collegeTuitionTesting['PredictedInstitution_Type_Binary'].apply(lambda x: 1 if x > 0.5 else 0)
```

```python
collegeTuitionTesting.head()
```

```python
# 25. Generate a confusion matrix of predictions
pd.crosstab(collegeTuitionTesting['Institution_Type_Binary'], collegeTuitionTesting['PredictedInstitution_Type_Binary'])
```

```python
# 26. Calculate the false positive rate for the model.
#     (Hint: False Positive Rate = False Positives / (False Positives + True Negatives))
falsePositiveRate = 1 - (pd.crosstab(collegeTuitionTesting['Institution_Type_Binary'], collegeTuitionTesting['PredictedInstitution_Type_Binary'])[0][0] / (pd.crosstab(collegeTuitionTesting['Institution_Type_Binary'], collegeTuitionTesting['PredictedInstitution_Type_Binary'])[0][0] + pd.crosstab(collegeTuitionTesting['Institution_Type_Binary'], collegeTuitionTesting['PredictedInstitution_Type_Binary'])[0][1]))
falsePositiveRate

#計算結果是0，這意味著模型沒有產生任何假陽性結果。這與先前的混淆矩陣一致，該矩陣顯示模型沒有將任何非私立學院錯誤地預測為私立學院。

#最後，代碼中的 falsePositiveRate 變量的值是0.0，這表明假陽性率計算結果為0。這與上述的分析相符合，因為從混淆矩陣看到沒有假陽性發生。
```

```python
# 27. Calculate the false negative rate for the model.
#     (Hint: False Negative Rate = False Negatives / (False Negatives + True Positives))
falseNegativeRate = 1 - (pd.crosstab(collegeTuitionTesting['Institution_Type_Binary'], collegeTuitionTesting['PredictedInstitution_Type_Binary'])[1][1] / (pd.crosstab(collegeTuitionTesting['Institution_Type_Binary'], collegeTuitionTesting['PredictedInstitution_Type_Binary'])[1][1] + pd.crosstab(collegeTuitionTesting['Institution_Type_Binary'], collegeTuitionTesting['PredictedInstitution_Type_Binary'])[1][0]))
falseNegativeRate

```

```python
# 28. Calculate prediction accuracy for the model and display it to the console.
#     (Hint: Prediction Accuracy = (True Positives + True Negatives) / Total)
predictionAccuracy = (pd.crosstab(collegeTuitionTesting['Institution_Type_Binary'], collegeTuitionTesting['PredictedInstitution_Type_Binary'])[0][0] + pd.crosstab(collegeTuitionTesting['Institution_Type_Binary'], collegeTuitionTesting['PredictedInstitution_Type_Binary'])[1][1]) / len(collegeTuitionTesting)
predictionAccuracy
```
